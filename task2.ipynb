```python
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.integrate import quad
import numpy as np
```

```python
mu = 110
sigma = 10
c = 10
p = 15
pl = 5
cs = np.linspace(0, 5, 1000)
values = (p - c) / (p + cs - pl)
qstar = norm.ppf(values, mu, sigma)
```

```python
plt.plot(cs, qstar)
plt.xlabel("Shipping cost (cs)")
plt.ylabel("Optimal order quantity (Q*)")
plt.title("Sensitivity of optimal order quantity")
```

```python
expected_profit = []
for i, j in zip(qstar, cs):
    value = (p - c) * i - (p + j - pl) * quad(lambda y: norm.cdf(y, loc=mu, scale=sigma), -np.inf, i)[0]
    expected_profit.append(value)
```

```python
plt.plot(cs, expected_profit)
plt.xlabel("Shipping cost (cs)")
plt.ylabel("Expected profit")
plt.title("Sensitivity of expected profit")
```
